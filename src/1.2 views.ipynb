```python
%sql SHOW TABLES
```

```python
%sql
create view apple_phones
as select * from smartphone where brand = 'Apple';
```

```python
%sql
select * from apple_phones;
```

```python
%sql

drop view if exists temp_v_brand_phones;

create temporary view temp_v_brand_phones
as select distinct brand from smartphone;

select * from temp_v_brand_phones;
```

```python
%sql

drop view if exists glob_temp_v_latest_phones;

create global temporary view glob_temp_v_latest_phones
as select * from smartphone where year > 2020 order by year desc;

select * from glob_temp_v_latest_phones;

--  GLOBAL TEMPORARY VIEW is not supported on serverless compute, thanks databricks free edition who only offers serverless compute :)

select * from global_temp.glob_temp_v_latest_phones;
```
