### internal vs external tables

```python
%sql
-- use the catalog again, in case of...
-- no hive metastore in this environment but following the exemple. So there is no "location" for a classic table
USE CATALOG joe;

CREATE TABLE managed_default (width INT, lenght INT, height INT);

INSERT INTO managed_default values(3 INT, 2, 1)
```

```python
%sql DESCRIBE EXTENDED managed_default
```

```python
%sql
-- This code creates a managed table in Unity Catalog. If you need to reference external data, you must first create a connection and then a foreign catalog, as described in the documentation.

-- Please note that you must use a Unity Catalog catalog and schema, and not the dbfs file scheme.

-- user add : this code could work with s3 and not dbfs

CREATE EXTERNAL TABLE external_default
(
  width INT
   , lenght INT
   , height INT
)
LOCATION 'dbfs:/Workspace/Users/nicolas.louis.jandot@gmail.com/Demo/external_default'
```

Note: The purpose of this exercise was to delete the tables and observe the presence of the files (delta logs, parquet files) with the external table.

```python
%sql
CREATE SCHEMA new_default
```

```python
%sql DESCRIBE DATABASE EXTENDED new_default
```

```python
%sql
use joe.new_default;
-- USE new_default;

CREATE TABLE IF NOT EXISTS managed_new_default (width INT, lenght INT, height INT);

INSERT INTO managed_new_default values(3 INT, 2, 1);
```
