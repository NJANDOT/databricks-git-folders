```python
%sql
CREATE OR REPLACE TEMPORARY VIEW tv_customers_update AS select * from json.`/Volumes/datademo/default/bookstore_dataset/customers-json-new`;

merge into customers c using tv_customers_update u on c.customer_id = u.customer_id when matched AND c.email is NULL and u.email IS NOT NULL THEN 
  UPDATE SET email = u.email, updated = u.updated
WHEN NOT MATCHED THEN INSERT *
```

```python
%sql
CREATE OR REPLACE TEMP VIEW books_updates AS
SELECT * FROM read_files(
    '/Volumes/datademo/default/bookstore_dataset/books-csv-new',
    format => 'csv',
    header => 'true',
    delimiter => ';');

SELECT * FROM books_updates
```

```python
%sql

merge into books b
using books_updates u
on b.book_id = u.book_id AND b.title = u.title
when not matched and u.category = 'Computer Science' THEN insert *
```
